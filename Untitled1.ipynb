```python
#1a)
reset()
t,w0,k=var('t,w0,k')
assume(w0,'integer')
assume(w0>0)
x=function('x')(t)
edq=diff(x,t,2)+(w0^2)*x==0
show(edq)
```

```python
#b)
desolve(edq,x,ivar=t)
v0=var('v0')
x0=var('x0')
sol=desolve(edq,x,ivar=t,ics=[0,x0,v0])
show(sol)
```

```python
#1c)
g=9.81
L=0.3924
t=var('t')
w0,x0,v0=var('w0,x0,v0')
assume(w0,'integer')
assume(w0>0)
x0=0.15
v0=0
w0=sqrt(g/L)
R=sqrt((x0^2)+((v0/w0)^2))
T=(2*pi/w0)
d=0
x=function('x')(t)
x=R*cos(w0*t-d)
show(x.full_simplify())
```

```python
plot(x,t,0,T)
```
